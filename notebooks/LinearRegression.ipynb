## What is Linear Regression?

Linear Regression is a fundamental supervised learning algorithm used to model the **relationship between one or more features and a continuous target**.

![LinearRegression](../diagrams/linear-regression.png)

It tries to fit the **best straight line** (or hyperplane) through the data points.

The general formula:

$
\hat{y} = w \cdot x + b
$

- `w`: weight or slope (how much y changes with x)
- `b`: bias or intercept (y-value when x = 0)

##  Why Use Linear Regression?

Because it's:
- Easy to interpret
- Fast to compute
- Works well for linearly correlated data
- Great for learning before moving to complex models

Used when:
- You want a baseline model
- You care about explaining the relationship between variables
- Your data has a roughly **linear trend**

## Who Uses It?

- Business analysts forecasting sales
- Economists predicting growth
- Data scientists estimating trends
- Students learning the basics of machine learning!

## Where is it Used?

Linear regression is applied in many domains:

- Stock price prediction
- House value estimation
- Sales forecasting
- Risk scoring in healthcare
- Supply/demand estimation

## When Should You Use It?

Use Linear Regression when:
- The relationship is (mostly) linear
- You want interpretable models
- Your features are not highly correlated
- You don't have major outliers

**Avoid it when**:
- Relationships are non-linear
- Multicollinearity is strong
- Outliers dominate the dataset

## How Does It Work?

Linear Regression minimizes the **difference between the predicted values and the actual values** using a loss function.

The most common loss is:

$
\text{MSE} = \frac{1}{n} \sum (y_{true} - y_{pred})^2
$

It uses an optimization algorithm (like gradient descent) to find the best `w` and `b` values that minimize this error.

# Trying out with random data generation

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 5 + 3 * X.squeeze() + np.random.randn(100)

plt.scatter(X, y, alpha=0.6)
plt.title("Generated Linear Data: y = 5 + 3x + noise")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

## Let's Train Our Custom Linear Regression Model

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))
```

```python
from algorithms.linear_regression import LinearRegression

model = LinearRegression(learning_rate=0.01, n_iterations=1000)
model.fit(X, y)
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, label="Actual", alpha=0.5)
plt.plot(X, y_pred, color="red", label="Prediction (Custom)", linewidth=2)
plt.title("ðŸ“‰ Custom Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

## How Good is Our Fit? Let's Evaluate!
Weâ€™ll use the following metrics:

- RÂ² (Coefficient of Determination)
- MSE (Mean Squared Error)
- MAE (Mean Absolute Error)
- Huber Loss (Robust to outliers)

```python
print("RÂ² Score:", model.score(X, y, metric="r2"))
print("MSE     :", model.score(X, y, metric="mse"))
print("MAE     :", model.score(X, y, metric="mae"))
print("Huber   :", model.score(X, y, metric="huber", delta=1.0))
```

## Linear Regression Variants

### Ridge Regression (L2 Regularization)
- Adds penalty for large weights
- Helps with multicollinearity

### Lasso Regression (L1 Regularization)
- Can shrink coefficients to zero (good for feature selection)

### ElasticNet
- Combines both L1 and L2


## When NOT to Use Linear Regression

- Data has complex non-linear patterns  
- You care more about accuracy than explainability  
- Features are strongly correlated with each other  
- Your data contains extreme outliers or noise

## Summary

- Linear Regression is the foundation of many machine learning models.
- Itâ€™s easy to understand, explain, and fast to compute.
- Use it for **continuous value prediction** when your data is linearly related.
